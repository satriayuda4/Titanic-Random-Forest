## Data Test Imputation

```python
df_test["Age"].fillna(df_test.Age.mean(), inplace=True)
df_test.tail()
```

```python
df_test["Fare"].fillna(df_test.Fare.mean(), inplace=True)
df_test.tail()
```

## Data Train Numeric Variables Analysis

```python
numeric_variables = list(df_train.dtypes[df_train.dtypes != "object"].index)
df_train[numeric_variables].head() 
```

```python
df_train[numeric_variables].head()
```

## Random Forest Algorithm for Numeric Variables

```python
#Libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
y = df_train['Survived']
y.head()
```

```python
y = df_train.pop('Survived')
y.head()
```

```python
numeric_variables.remove('Survived')
```

```python
x = df_train[numeric_variables]
x
```

## Random Forest Training

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(x, y)
```

```python
a = model.predict(df_train[numeric_variables])
print ("Train Accuracy :", accuracy_score(y,a))
```

## Data Test Numeric Variable Analysis

```python
df_test[numeric_variables].head()
```

```python
df_test[numeric_variables].isnull().sum()
```

## Random Forest Prediction on Data Test

```python
y_pred = model.predict(df_test[numeric_variables])
y_pred
```

## Submission

```python
submission = pd.DataFrame({
    "PassengerId": df_test["PassengerId"],
    "Survived": y_pred
})

submission.to_csv('titanicPresentation.csv', index=False, sep='\t')
submission.head()
```
